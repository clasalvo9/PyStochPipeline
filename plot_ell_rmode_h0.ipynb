```python
#plottare epsilon

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# === Matplotlib config ===
matplotlib.rcParams.update({
    'text.usetex': True,
    'font.size': 20,
    'axes.linewidth': 1.2,
    'font.family': 'serif',
    'figure.figsize': (9, 7)
})

# === Costanti fisiche ===
I = 1e38  # momento d'inerzia NS [kg m^2]
kpc_to_m = 3.086e19
pi = np.pi
G = 6.67430e-11  # m^3 kg^-1 s^-2
c = 299792458    # m/s

# === Distanze min/max per ogni target [kpc] ===
dist_dict = {
    "Vela": (0.5, 1.3),
    "G347": (0.9, 1.1),
    "SN1987A": (50, 50),
    "CasA": (3.3, 3.7),
}

# === Colori ===
colors = {
    "Vela": 'tab:blue',
    "G347": 'tab:green',
    "SN1987A": 'tab:orange',
    "CasA": 'tab:red'
}

# === Etichette formattate per la legenda ===
target_labels = {
    "G347": "G347.3$-$0.5",
    "Vela": "G266.2$-$1.2 (Vela Jr)",
    "CasA": "G111.7$-$2.1 (Cas A)",
    "SN1987A": "G279.7$-$31.9 (SNR 1987A)"
}

# === Carica i dati ===
data = np.loadtxt("N_0.txt", skiprows=1)
#f_gw = f0 #data[:, 0]  # frequenza GW [Hz]

# === Upper limits h0 ===
#h0_targets = {
#    "Vela": data[:, 1],
#    "G347": data[:, 2],
#    "SN1987A": data[:, 3],
#    "CasA": data[:, 4]
#}
h0_targets = {
    "Vela": sig0V,
    "G347": sig0G,
    "SN1987A": sig0N,
    "CasA": sig0C
}

f0_targets = {
    "Vela": f0V,
    "G347": f0G,
    "SN1987A": f0N,
    "CasA": f0C
}


# === Funzione per epsilon ===
def compute_epsilon_from_fgw(h0, d_kpc, fgw, I=1e38):
    d_m = d_kpc * kpc_to_m
    epsilon = (h0 * d_m * c*c*c*c) / (4 * np.pi*np.pi * G * I * fgw**2)
    return epsilon

# === Campionamento ogni 1 Hz per i puntini ===
#_, thin_indices = np.unique(np.round(f_gw), return_index=True)

# === Crea il grafico ===
plt.figure()

for target in h0_targets:
    h0 = h0_targets[target]
    f = f0_targets[target]
    #f = f_gw
    dmin, dmax = dist_dict[target]

    eps_min = compute_epsilon_from_fgw(h0, dmin, f)
    eps_max = compute_epsilon_from_fgw(h0, dmax, f)

    # Campionamento ridotto solo per i puntini
    #f_thin = f[thin_indices]
    #eps_min_thin = eps_min[thin_indices]

    # Zona sfumata completa (tutti i punti)
    if dmin != dmax:
        plt.fill_between(f, eps_min, eps_max, color=colors[target], alpha=0.3)

    # Puntini campionati
    plt.plot(f, eps_min, 'o', color=colors[target],
             markersize=4, label=target_labels[target])

plt.xlim(f[0], f[-1])
plt.yscale('log')
#plt.xscale('log')
plt.xlabel(r"$f$ (Hz)")
plt.ylabel(r"$\epsilon^{95\%}$")
plt.grid(True, which='both', linestyle='-', linewidth=0.5)
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig(fr"ellipticity_all_log.png", dpi=300)
plt.close()
```

```python
#plottare rmodes

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# === Matplotlib config ===
matplotlib.rcParams.update({
    'text.usetex': True,
    'font.size': 20,
    'axes.linewidth': 1.2,
    'font.family': 'serif',
    'figure.figsize': (9, 7)
})

# === Costanti fisiche ===
I = 1e38  # momento d'inerzia NS [kg m^2]
kpc_to_m = 3.086e19
pi = np.pi
G = 6.67430e-11  # m^3 kg^-1 s^-2
c = 299792458    # m/s

# === Distanze min/max per ogni target [kpc] ===
dist_dict = {
    "Vela": (0.5, 1.3),
    "G347": (0.9, 1.1),
    "SN1987A": (50, 50),
    "CasA": (3.3, 3.7),
}

# === Colori ===
colors = {
    "Vela": 'tab:blue',
    "G347": 'tab:green',
    "SN1987A": 'tab:orange',
    "CasA": 'tab:red'
}

# === Etichette formattate per la legenda ===
target_labels = {
    "G347": "G347.3$-$0.5",
    "Vela": "G266.2$-$1.2 (Vela Jr)",
    "CasA": "G111.7$-$2.1 (Cas A)",
    "SN1987A": "G279.7$-$31.9 (SNR 1987A)"
}


# === Carica i dati ===
data = np.loadtxt("N_0.txt", skiprows=1)
#f_gw = f0 #data[:, 0]  # frequenza GW [Hz]

# === Upper limits h0 ===
#h0_targets = {
#    "Vela": data[:, 1],
#    "G347": data[:, 2],
#    "SN1987A": data[:, 3],
#    "CasA": data[:, 4]
#}
h0_targets = {
    "Vela": sig0V,
    "G347": sig0G,
    "SN1987A": sig0N,
    "CasA": sig0C
}

f0_targets = {
    "Vela": f0V,
    "G347": f0G,
    "SN1987A": f0N,
    "CasA": f0C
}


# === Funzione per epsilon ===

def compute_rmodes_from_fgw(h0,  d_kpc, fgw):
    kpc_in_meters = 3.085677581e19  # 1 kpc in metri
    D = d_kpc * kpc_in_meters
    alpha = 0.028 * (h0 / 1e-24) * (100 / fgw) ** 3 * (D / kpc_in_meters)
    return alpha

# === Campionamento ogni 1 Hz per i puntini ===
#_, thin_indices = np.unique(np.round(f_gw), return_index=True)

# === Crea il grafico ===
plt.figure()

for target in h0_targets:
    h0 = h0_targets[target]
    f = f0_targets[target]
    #f = f_gw
    dmin, dmax = dist_dict[target]

    eps_min = compute_rmodes_from_fgw(h0, dmin, f)
    eps_max = compute_rmodes_from_fgw(h0, dmax, f)

    # Campionamento ridotto solo per i puntini
    #f_thin = f[thin_indices]
    #eps_min_thin = eps_min[thin_indices]

    # Zona sfumata completa (tutti i punti)
    if dmin != dmax:
        plt.fill_between(f, eps_min, eps_max, color=colors[target], alpha=0.3)

    # Puntini campionati
    plt.plot(f, eps_min, 'o', color=colors[target],
             markersize=4, label=target_labels[target])

plt.xlim(f[0], f[-1])
plt.yscale('log')
#plt.xscale('log')
plt.xlabel(r"$f$ (Hz)")
plt.ylabel(r"$\alpha^{95\%}$")
plt.grid(True, which='both', linestyle='-', linewidth=0.5)
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig(fr"rmodes_alltargets_log.png", dpi=300)
plt.close()
```

```python
from matplotlib import pyplot as plt
import numpy
import scipy.io

def get_running_median(h0, freqs):
    running_median_h0 = numpy.zeros(h0.size) * numpy.nan
    for ii, ff in enumerate(freqs):
        idxs = (freqs > (ff-0.5)) * (freqs < (ff + 0.5))
        running_median_h0[ii] = numpy.nanmedian(h0[idxs])
    return running_median_h0

def get_best_freq_and_h0(h0, freqs):
    med = get_running_median(h0, freqs)
    best_freq = freqs[numpy.nanargmin(med)]
    best_h0 = numpy.nanmin(med)
    print('{0:4.6e} at {1:4.6f}'.format(best_h0, best_freq))
    return best_h0, best_freq

#parent_folder  = "/home/iuri.larosa/O4/output/O4a0l/"
points = numpy.array(["sn1987a", "G347", "CA", "Vela"])
R='O4'

point = 0
print(points[point])
struct = scipy.io.loadmat( points[point]+"_ul/" + R + "/0_final_results.mat", squeeze_me=True)

med = get_running_median(struct['ul'], struct['f'])
#best = get_best_freq_and_h0(struct['ul'], struct['f'])


plt.loglog(struct['f'], struct['ul'], marker='o', linestyle='None',markersize='3', color='C0', label="Best sn1987a")  # Rosso per PyStoch
plt.loglog(struct['f'], med, marker='o', linestyle='None',markersize='3', color='C1', label="Average sn1987a")  # Rosso per PyStoch
plt.xlabel("Frequency [Hz]", fontsize=14)
plt.ylabel(r"$h_0^{95\%}$", fontsize=14)

# Tick font più grande
plt.tick_params(axis='both', which='major', labelsize=12)

# Legenda con font più grande
plt.legend(fontsize=12)

plt.xlim(20,1726)
#plt.ylim(1e-25,1e-23)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

#print(med)
get_best_freq_and_h0(struct['ul'], struct['f'])
```
